# Rating Prediction Project

#### Data description

We have a client who has a website where people write different reviews for technical products.
Now they are adding a new feature to their website i.e. The reviewer will have to add stars(rating)
as well with the review. The rating is out 5 stars and it only has 5 options available 1 star, 2 stars,
3 stars, 4 stars, 5 stars. Now they want to predict ratings for the reviews which were written in the
past and they don’t have a rating. So, we have to build an application which can predict the rating
by seeing the review.
You have to scrape at least 20000 rows of data.
In this section you need to scrape the reviews of different laptops, Phones, Headphones, smart
watches, Professional Cameras, Printers, Monitors, Home theater, Router from different ecommerce websites.
Basically, we need these columns:  
1) reviews of the product.   
2) rating of the product.

After collecting the data, you need to build a machine learning model. Before model building we need to do
all data preprocessing steps involving NLP. Then try different models with different hyper parameters
and select the best model.

### Scraping Reviews for Laptops

```python
# Giving address to the driver

driver.get('https://www.flipkart.com/')
```

```python
# Finding element for the search bar and giving value inside it.

search_box= driver.find_element_by_xpath("//div[@class='_3OO5Xc']/input")
search_box.send_keys('laptop')
time.sleep(2)

# Clicking on the search buton

driver.find_element_by_xpath("//button[@class='L0Z3Pu']").click()
```

```python
# Extracting page urls of all the products.

pages=[]
page= driver.find_elements_by_xpath("//a[@class='_1fQZEK']")
for i in page:
    pages.append(i.get_attribute('href'))
```

```python
# creating lists to store data.

rating=[]
review=[]

# Scraping the Ratings and reviews.

for i in pages:
    driver.get(i)
    time.sleep(3)
    try:
        # Clicking on view all review.
        a=driver.find_element_by_xpath("//div[@class='_3UAT2v _16PBlm']")
        a.click()
        time.sleep(3)
        
        # Clicking on the read more options to display entire text.
        for j in driver.find_elements_by_xpath("//span[@class='_1BWGvX']"):
            j.click()
        
        # Scraping data.
        for k in range(0,12):
            for t in driver.find_elements_by_xpath("//div[@class='col _2wzgFH K0kLPL']/div[1]/div"):
                rating.append(t.text)
            for r in driver.find_elements_by_xpath("//div[@class='t-ZTKy']/div/div"):
                review.append(r.text)
                
            # Clicking o the next button.
            driver.find_element_by_xpath("//span[contains(text(),'Next')]").click()
            time.sleep(3)
                
    except NoSuchElementException:
        pass
```

```python
# Creating dataframe for the data.

laptop1= pd.DataFrame({"Review":review,"Rating":rating})
laptop1
```

```python
# Creating lists to store data

rev5=[]
rev4=[]
rev3=[]
rev2=[]
rev1=[]

# Scraping the reviews.

for i in pages:
    driver.get(i)
    time.sleep(3)
    
    # Scraping reviews for 5 stars ratings.
    try:
        a=driver.find_element_by_xpath("//a[@class='a-link-normal 5star']")
        a.click()
        time.sleep(3)
        for j in range(0,8):
            for r in driver.find_elements_by_xpath("//span[@class='a-size-base review-text review-text-content']/span[1]"):
                rev5.append(r.text)
            driver.find_element_by_xpath("//a[contains(text(),'Next page')]").click()
            time.sleep(3)
                
    except NoSuchElementException:
        pass
    
    # Scraping reviews for 4 stars ratings.
    try:
        a=driver.find_element_by_xpath("//a[@class='a-link-normal 4star']")
        a.click()
        time.sleep(3)
        for j in range(0,8):
            for r in driver.find_elements_by_xpath("//span[@class='a-size-base review-text review-text-content']/span[1]"):
                rev4.append(r.text)
            driver.find_element_by_xpath("//a[contains(text(),'Next page')]").click()
            time.sleep(3)
        
    except NoSuchElementException:
        pass
    
    # Scraping reviews for 3 stars ratings.
    try:
        a=driver.find_element_by_xpath("//a[@class='a-link-normal 3star']")
        a.click()
        time.sleep(3)
        for j in range(0,8):
            for r in driver.find_elements_by_xpath("//span[@class='a-size-base review-text review-text-content']/span[1]"):
                rev3.append(r.text)
            driver.find_element_by_xpath("//a[contains(text(),'Next page')]").click()
            time.sleep(3)
        
    except NoSuchElementException:
        pass
    
    # Scraping reviews for 2 stars ratings.
    try:
        a=driver.find_element_by_xpath("//a[@class='a-link-normal 2star']")
        a.click()
        time.sleep(3)
        for j in range(0,8):
            for r in driver.find_elements_by_xpath("//span[@class='a-size-base review-text review-text-content']/span[1]"):
                rev2.append(r.text)
            driver.find_element_by_xpath("//a[contains(text(),'Next page')]").click()
            time.sleep(3)
        
    except NoSuchElementException:
        pass
    
    # Scraping reviews for 1 star ratings.
    try:
        a=driver.find_element_by_xpath("//a[@class='a-link-normal 1star']")
        a.click()
        time.sleep(3)
        for j in range(0,8):
            for r in driver.find_elements_by_xpath("//span[@class='a-size-base review-text review-text-content']/span[1]"):
                rev1.append(r.text)
            driver.find_element_by_xpath("//a[contains(text(),'Next page')]").click()
            time.sleep(3)
        
    except NoSuchElementException:
        pass
```

Here we are extracting the reviews one by one for all ratings. this is because there is no rating displayed in text in the website.

```python
# Creating dataframes and concatenating them.

r1= pd.DataFrame({"Review":rev1,"Rating":1})
r2= pd.DataFrame({"Review":rev2,"Rating":2})
r3= pd.DataFrame({"Review":rev3,"Rating":3})
r4= pd.DataFrame({"Review":rev4,"Rating":4})
r5= pd.DataFrame({"Review":rev5,"Rating":5})

laptop2=pd.concat([r1, r2,r3,r4,r5], axis=0)
laptop2
```

```python
# Creating lists to store data

rev5=[]
rev4=[]
rev3=[]
rev2=[]
rev1=[]

# Scraping the reviews.

for i in pages:
    driver.get(i)
    time.sleep(3)
    
    # Scraping reviews for 5 stars ratings.
    try:
        a=driver.find_element_by_xpath("//a[@class='a-link-normal 5star']")
        a.click()
        time.sleep(3)
        for j in range(0,4):
            for r in driver.find_elements_by_xpath("//span[@class='a-size-base review-text review-text-content']/span[1]"):
                rev5.append(r.text)
            driver.find_element_by_xpath("//a[contains(text(),'Next page')]").click()
            time.sleep(3)
                
    except NoSuchElementException:
        pass
    
    # Scraping reviews for 4 stars ratings.
    try:
        a=driver.find_element_by_xpath("//a[@class='a-link-normal 4star']")
        a.click()
        time.sleep(3)
        for j in range(0,4):
            for r in driver.find_elements_by_xpath("//span[@class='a-size-base review-text review-text-content']/span"):
                rev4.append(r.text)
            driver.find_element_by_xpath("//a[contains(text(),'Next page')]").click()
            time.sleep(3)
        
    except NoSuchElementException:
        pass
    
    # Scraping reviews for 3 stars ratings.
    try:
        a=driver.find_element_by_xpath("//a[@class='a-link-normal 3star']")
        a.click()
        time.sleep(3)
        for j in range(0,8):
            for r in driver.find_elements_by_xpath("//span[@class='a-size-base review-text review-text-content']/span"):
                rev3.append(r.text)
            driver.find_element_by_xpath("//a[contains(text(),'Next page')]").click()
            time.sleep(3)
        
    except NoSuchElementException:
        pass
    
    # Scraping reviews for 2 stars ratings.
    try:
        a=driver.find_element_by_xpath("//a[@class='a-link-normal 2star']")
        a.click()
        time.sleep(3)
        for j in range(0,8):
            for r in driver.find_elements_by_xpath("//span[@class='a-size-base review-text review-text-content']/span"):
                rev2.append(r.text)
            driver.find_element_by_xpath("//a[contains(text(),'Next page')]").click()
            time.sleep(3)
        
    except NoSuchElementException:
        pass
    
    # Scraping reviews for 1 star ratings.
    try:
        a=driver.find_element_by_xpath("//a[@class='a-link-normal 1star']")
        a.click()
        time.sleep(3)
        for j in range(0,5):
            for r in driver.find_elements_by_xpath("//span[@class='a-size-base review-text review-text-content']/span"):
                rev1.append(r.text)
            driver.find_element_by_xpath("//a[contains(text(),'Next page')]").click()
            time.sleep(3)
        
    except NoSuchElementException:
        pass
```
